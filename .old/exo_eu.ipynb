```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import re
```

```python
t_eu = pd.read_csv('exoplanet.eu_catalog.csv')
# t_eu.set_index('# name', inplace=True)
# print(t_eu)
t_eu
```

```python
re_2mass = re.compile(r"^\s*(2MASS(\s|-)((J|H|K)?\d+(\+|-)\d+))")
re_bd = re.compile(r"^\s*(BD(\s|\+|-)(\d+\s\d+))")
re_hd = re.compile(r"^\s*(HD\s(\d+))")
re_hip = re.compile(r"^\s*(HIP\s(\d+))")
re_gj = re.compile(r"^\s*(GJ\s(\d+))")
re_kic = re.compile(r"^\s*(KIC\s(\d+))")
re_koi = re.compile(r"^\s*(KOI-(\d+))")
re_tic = re.compile(r"^\s*(TIC\s(\d+))")
re_toi = re.compile(r"^\s*(TOI-(\d+))")
re_tyc = re.compile(r"^\s*(TYC\s(\d+(-\d+)*))")
def alt_name_cols(row):
    
    new_row = row.copy(deep=False)
    
    s_alt_names_str = str(row['star_alternate_names'])
    new_row['SNAME_2MASS'] = ""
    new_row['SNAME_BD'] = ""
    new_row['SNAME_HD'] = ""
    new_row['SNAME_HIP'] = ""
    new_row['SNAME_GJ'] = ""
    new_row['SNAME_KIC'] = ""
    new_row['SNAME_KOI'] = ""
    new_row['SNAME_TIC'] = ""
    new_row['SNAME_TOI'] = ""
    new_row['SNAME_TYC'] = ""
    
    s_name_list = s_alt_names_str.split(",")
    
    for s_name in reversed(s_name_list):
        m_2mass = re.search(re_2mass, s_name)
        if m_2mass is not None:
            new_row['SNAME_2MASS'] = m_2mass.group(3)
            continue
        
        m_bd = re.search(re_bd, s_name)
        if m_bd is not None:
            new_row['SNAME_BD'] = m_bd.group(3)
            continue
        
        m_hd = re.search(re_hd, s_name)
        if m_hd is not None:
            new_row['SNAME_HD'] = m_hd.group(2)
            continue
        
        m_hip = re.search(re_hip, s_name)
        if m_hip is not None:
            new_row['SNAME_HIP'] = m_hip.group(2)
            continue
                
        m_gj = re.search(re_gj, s_name)
        if m_gj is not None:
            new_row['SNAME_GJ'] = m_gj.group(2)
            continue
        
        m_kic = re.search(re_kic, s_name)
        if m_kic is not None:
            new_row['SNAME_KIC'] = m_kic.group(2)
            continue
        
        m_koi = re.search(re_koi, s_name)
        if m_koi is not None:
            new_row['SNAME_KOI'] = m_koi.group(2)
            continue
        
        m_tic = re.search(re_tic, s_name)
        if m_tic is not None:
            new_row['SNAME_TIC'] = m_tic.group(2)
            continue
        
        m_toi = re.search(re_toi, s_name)
        if m_toi is not None:
            new_row['SNAME_TOI'] = m_toi.group(2)
            continue
        
        m_tyc = re.search(re_tyc, s_name)
        if m_tyc is not None:
            new_row['SNAME_TYC'] = m_tyc.group(2)
            continue
            
    return new_row
```

```python
newnames = t_eu.apply(alt_name_cols, axis=1)
```
