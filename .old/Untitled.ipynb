```python
p_kic = re.compile(r'^KIC (\d+)$')
p_koi = re.compile(r'^KOI-(\d+)$')
p_kep = re.compile(r'^Kepler-(\d+)$')

def kepler_names(hostname):
    m_kic = p_kic.match(hostname)
    if m_kic is not None:
        kic = m_kic.group(1)
        row = nasa_kep[nasa_kep.kepid == kic]
        return row
    
    m_kep = p_kep.match(hostname)
    if m_kep is not None:
        kep = m_kep.group(0)
        row = nasa_kep[nasa_kep.st_name == kep]
        return row
    
```
