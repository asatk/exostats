```python
# wright et al 2018 (eqn 5): valid for range 1.1 < V-K < 7.0
def taucVK(V,K):
    if np.fabs(V - K - 4.05) <= 2.95:
        return pow(10.,0.64 + 0.25 * (V - K))
    else:
        return np.nan

def RoVK(Prot, V, K):
    return np.fabs(Prot) / taucVK(V,K)
    
    
# wright et al 2018 (eqn 6): valid for range 0.08 < M/Msol < 1.36
#M in solar masses
def taucM(M):
    if np.fabs(M - 0.72) <= 0.64:
        return pow(10., 2.33 - 1.50 * M + 0.31 * pow(M, 2.)) 
    else:
        return np.nan
    
def RoM(Prot, M):
    return np.fabs(Prot) / taucM(M)

def RoAvg(RoVK, RoM):
    if not np.isnan(RoVK) and not np.isnan(RoM):
        return (RoVK + RoM) / 2.
    elif np.isnan(RoVK) and not np.isnan(RoM):
        return RoM
    elif not np.isnan(RoVK) and np.isnan(RoM):
        return RoVK
    else:
        return np.nan
```

```python
prot_data['RoVK'] = prot_data.apply(lambda x: RoVK(x['PROT'], x['sy_vmag'], x['sy_kmag']), axis=1)
prot_data['TaucVK'] = prot_data.apply(lambda x: taucVK(x['sy_vmag'], x['sy_kmag']), axis=1)
prot_data['RoM'] = prot_data.apply(lambda x: RoM(x['PROT'], x['st_mass']), axis=1)
prot_data['TaucM'] = prot_data.apply(lambda x: taucM(x['st_mass']), axis=1)
prot_data['RoAvg'] = prot_data.apply(lambda x: RoAvg(x['RoVK'], x['RoM']), axis=1)
```

```python
#choose which Ro is used in plots
prot_data['Ro'] = prot_data['RoM']
```

```python
col_list = ['name','pl_orbsmax','pl_orbeccen','pl_rade','pl_bmasse','pl_bmassprov','st_rad','RoVK','RoM','RoAvg', 'Ro']
data = prot_data[prot_data['Ro'].notnull() & prot_data['pl_orbsmax'].notnull() & prot_data['pl_orbeccen'].notnull()]
data = data[col_list].copy()
data
```

```python
data['rperi'] = data.apply(lambda x: x['pl_orbsmax'] if x['pl_orbeccen'] == 0.0 else x['pl_orbsmax'] * (1 - x['pl_orbeccen']), axis=1)
```

```python
x = np.linspace(0.,2.7,1000) # ro/ro sol x vals
prot_sol = 27.
ro_sol = 1.85
ra_sol = 0.1
r_sol = 1/200.
# all values scaled to solar values at maximum
# combining eqns: (7) Farrish 2019 and (2) Farrish 2021
def ra_schrijver(ro, rad):
    return np.real(pow(ro / ro_sol, -1.5 * -0.16) * pow(rad / r_sol, -2 * -0.16))
```

```python
alfven = data.apply(lambda x: x['rperi'] / (ra_sol * ra_schrijver(x['Ro'], x['st_rad'])), axis=1)
alfven_data = data.copy(deep=False)
alfven_data['orbit-alfven'] = alfven
alfven_data = alfven_data[alfven_data['st_rad'].notnull()]
#optional: remove 99th quantile and bad vals
alfven_data = alfven_data[(alfven_data['Ro'] < np.quantile(alfven_data['Ro'], 0.99)) & (alfven_data['orbit-alfven'] < np.quantile(alfven_data['orbit-alfven'], 0.99))]
alfven_data = alfven_data[(alfven_data['Ro'] > 0.0) & (alfven_data['orbit-alfven'] > 0.0)]
# alfven_data
```

```python
ylim = np.ceil(np.quantile(alfven, 0.99))
ylim
```

```python
alfven_habitable = pd.merge(habitable, alfven_data, on='name')
# alfven_habitable
```

```python
alfven_bad = alfven_data[alfven_data['orbit-alfven'] < 0.8]
alfven_maybe = alfven_data[(alfven_data['orbit-alfven'] > 0.8) & (alfven_data['orbit-alfven'] < 1.2)]
alfven_good = alfven_data[alfven_data['orbit-alfven'] > 1.2]
print([len(alfven_bad),len(alfven_maybe),len(alfven_good)])
```

```python
fig2 = plt.figure(figsize=(15,8))
ax2 = fig2.add_subplot(111)
ax2.scatter(alfven_bad['Ro'],alfven_bad['orbit-alfven'], color='#FF6666', s=75, label='Inside AS')
ax2.scatter(alfven_maybe['Ro'],alfven_maybe['orbit-alfven'], color='#DDBB44', s=75, label='Near AS')
ax2.scatter(alfven_good['Ro'],alfven_good['orbit-alfven'], color='#66FF66', s=75, label='Outside AS')
for idx, row in alfven_habitable.iterrows():
    xpt = row['Ro']
    ypt = row['orbit-alfven']
    lbl = row['name']
    ax2.scatter(xpt, ypt, color='#0000FF', s=250, marker='*')
    
ax2.set_title("Exoplanet Periastrons and RA Estimates",fontsize=25)
ax2.set_xlabel("Ro",fontsize=20)
ax2.set_ylabel("rp/RA",fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
ax2.grid(visible=True)
ax2.set_xlim([0.0,xlim])
ax2.set_ylim([0.0,ylim])
ax2.legend(loc=1,fontsize=15)
fig2.set_facecolor('white')
plt.axhline(y = 1.0, xmin = 0.00, xmax = np.max(data['Ro']), linestyle='--', color='#666666')
plt.text(xlim + .02, 0.98, 'AS', fontsize=15)
pass
```
